```python
# Count of missing values in each feautre 
training_data.isnull().sum()
```

```python
# Percentage of missing values in each feature

features_with_null = [feature for feature in training_data.columns if training_data[feature].isnull().sum()>1]

for feature in features_with_null:
    print(feature, round(training_data[feature].isnull().mean(), 2),'% missing values')
```

```python
# chosing the measure of central tendency for the target variable
res = stats.skew(training_data['SalePrice'])
print(res)
```

```python
#Mean, median and mode pf the target variable
mean = training_data['SalePrice'].mean()
print(f"mean of Sales Price is {round(mean,2)}")
median = training_data['SalePrice'].median()
print(f"median of Sales Price is {round(median,2)}")
mode = training_data['SalePrice'].mode()
print(f"mode of Sales price is {round(mode)}")
```

```python
min_price = training_data['SalePrice'].min()
max_price = training_data['SalePrice'].max()
price_range = (min_price,max_price)
price_range
```

```python
#plt.figure(figsize=(10,6))
sns.histplot(training_data['SalePrice'], kde=True)
plt.title('Distribution of Sale Price')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()
```

shape of the distribution => the distribution is right-skewed. 
interpretation 
most of the house prices are in a lower price range, with few with significanly high price and a peak around 140000
balanced or imbalanced ? => the distributon is imbalanced with most of the prices fall under the lower price range with few under the expensive price range



```python
# The relationship between the missing values and the target variable
for feature in features_with_null:
    data = training_data.copy()

    data[feature] = data[feature].isnull().astype(int)

    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

Numerical features

```python
numarical_features = [feature for feature in training_data.columns if training_data[feature].dtypes != 'O']

print("Number of numarical features",len(numarical_features))

training_data[numarical_features].head()
```

from the dataset we have 4 temporal veraibles (Eg: datetime veriables )

```python
temporal_features = [feature for feature in numarical_features if 'Yr' in feature or 'Year' in feature]
temporal_features
```

```python
# printing the unique values in the temporal features

for feature in temporal_features:
    print(feature, training_data[feature].unique())
```

checking the relationship between the temporal features and the target variable

```python
for feature in temporal_features:
    training_data.groupby(feature)['SalePrice'].median().plot()
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.title(f"{feature} vs Sale Price")
    plt.show()
```

Exploring the numarical features (discrete and continous)

```python
descrete_feature = [feature for feature in numarical_features if len(training_data[feature].unique())< 25 and feature not in temporal_features + ['Id']]
for feature in descrete_feature:
    print(feature, training_data[feature].nunique())
```

```python
training_data[descrete_feature].head()
```

The relationship between descrete features and the target variable

```python
for feature in descrete_feature:
    training_data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Sales Price')
    plt.title(f"{feature} vs Sales Price")
    plt.show()
```

The relationship between continuous features and the target variable

```python
continuous_features = [feature for feature in numarical_features if feature not in descrete_feature + temporal_features + ['Id']]
print("Number of continuous features", len(continuous_features))
```

The distribution of continuous features

```python
for feature in continuous_features:
    training_data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(f"Distribution of {feature}")
    plt.show()
```
